```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

### Load the data

```python
tasks_df = pd.read_csv("../assets/tasks.csv")
```

### Visualize the data

```python
fig, ax = plt.subplots(figsize=(5, 4))
sns.barplot(
    tasks_df.groupby("Activity")["Task ID"].count().reset_index().rename(columns={"Task ID": "# Tasks"}),
    x="Activity", y="# Tasks", hue="Activity",
    order=["Requirements", "Development", "Quality assurance", "Maintenance", "Management"],
    palette=["#17becf", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd"]
)
plt.xticks(rotation=45, ha="right", rotation_mode="anchor")
ax.set_xlabel("Activity   ", labelpad=-10) 
plt.rcParams['font.size'] = 13

plt.tight_layout()

plt.savefig("tasks_per_activity.pdf", format="pdf", bbox_inches="tight")

plt.show()
```
